```r
library(gbm)
gbm2 <- gbm(X9.00.AM~., data=mrh[,3:242], var.monotone=rep(0,239), distribution="gaussian", n.trees=100, shrinkage=0.1, interaction.depth=3, bag.fraction = .5, train.fraction = 1, n.minobsinnode = 10, cv.folds = 10, keep.data=TRUE, verbose=FALSE)
```

```r
best.iter <- gbm.perf(gbm2,method="cv");best.iter
sqrt(gbm2$cv.error[best.iter]) #CV error SD
1-gbm2$cv.error[best.iter]/var(mrh$X9.00.AM)  #CV r^2
yhat<-predict(gbm2,mrh, n.trees = best.iter); sd(mrh$X9.00.AM-yhat)  #training error SD
##
summary(gbm2,n.trees=best.iter)  # based on the optimal number of trees
```

```r
network <- matrix(mrh$station[1:134])
network1 <- matrix(nrow=134,ncol=1)
trafficflow <- matrix(mrh$station[1:134])
trafficflow1 <- matrix(nrow=134,ncol=1)
k=1
library(randomForest)
for (i in 1:134){
    for (n in 1:134){
   
    if(k>=124){
        k= 1
        s=k+10
    } 
    else{
        k=k+1
        s=k+10
    }
    rForest1 <- randomForest(X9.00.AM~., data=mrh[k:s,3:124], mtry=3, ntree = 500, nodesize = 3, importance = TRUE)
    imp<-data.frame(t(importance(rForest1)))
    imp<-imp[2,]
    t<-names(which.max(imp))
    network1[n] <- c(t)
    trafficflow1[n]<-mrh[n,which(names(mrh)==t)]
    }
    network<-cbind(network,network1)
    trafficflow<-cbind(trafficflow,trafficflow1)
}
write.csv(network，file="D:/northwestern courses/CIV480/research/network.csv")
write.csv(trafficflow，file="D:/northwestern courses/CIV480/research/trafficflow.csv")
```
